```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
cancer = load_breast_cancer()
print("Target names: ", cancer.target_names)
print("\nFeature names: ", cancer.feature_names)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

X = cancer.data[:, :2]
y = cancer.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear', C=500.0)  # High C value for hard margin
model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score
from mlxtend.plotting import plot_decision_regions

y_pred = model.predict(X_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision: {precision_score(y_test, y_pred):.2f}')
print(f'Recall: {recall_score(y_test, y_pred):.2f}')

# Visualize the decision boundary, training points, and support vectors
plt.figure(figsize=(10, 6))
plot_decision_regions(X_train, y_train, clf=model, legend=2)
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k', label='Support Vectors')

plt.title('SVM Decision Boundary with Linear Kernel')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend(loc='upper left')
plt.show()
```
